## **3. Medidas descriptivas**



## **3.1 Medidas de tendencia central**

Son medidas que nos ayudan a resumir una gran cantidad de informaci√≥n en un solo numero

## **Media:**
Es el promedio de todos los datos, puede ser susceptible a valores at√≠picos

**Media** = $$ \frac{1}{N} \sum_{i=1}^N x_i  $$

## **Mediana:**
Es el dato central es decir tiene la misa cantidad de datos a su izquierda y derecha, no es lo mismo que la media

### **Mediana(n=impar)** = $$ x_{(n+1)/2}^{\text{ordered}} $$

### **Mediana(n=par)** = $$ \frac{x_{(n/2)}^{\text{ordered}} + x_{(n/2)+1}^{\text{ordered}}}{2} $$

## **Moda:**
Es el dato que mas se repite, la moda no aplica para datos num√©ricos continuos

### **Moda** = $$ x_k $$ donde $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$

# 4.1 Tablas y gr√°ficas para una variable continua: para esta secci√≥n elegimos los trabajos completados y las ganancias


```python
#Tabla de frecuencias con los intervalos autom√°ticos

def tabla_frecuencia(datos):
    columna = datos['Job_Completed']  # Seleccionar la columna de inter√©s
    n = len(columna)  # Cantidad de datos
    k = math.ceil(1 + math.log2(n))  # Regla de Sturges
    min_val, max_val = min(columna), max(columna)
    rango = max_val - min_val
    amplitud = rango / k  # Amplitud de los intervalos

    # Definir los intervalos
    bins = np.linspace(min_val, max_val, k + 1)

    # Calcular frecuencias
    frecuencias, _ = np.histogram(columna, bins)

    # Calcular frecuencias acumuladas
    frec_acumulada = np.cumsum(frecuencias)

    # Calcular frecuencia relativa
    frec_relativa = frecuencias / n

    # Calcular frecuencia relativa acumulada
    frec_rel_acumulada = np.cumsum(frec_relativa)

    # Crear tabla de frecuencias
    tabla = pd.DataFrame({
        'Intervalo': [f'[{bins[i]:.2f}, {bins[i+1]:.2f})' for i in range(len(bins)-1)],
        'Frecuencia': frecuencias,
        'Frec. Acumulada': frec_acumulada,
        'Frec. Relativa': frec_relativa,
        'Frec. Rel. Acumulada': frec_rel_acumulada
    })

    return tabla

# Aplicar funci√≥n al DataFrame
resultado = tabla_frecuencia(datos)
print(resultado)
```
